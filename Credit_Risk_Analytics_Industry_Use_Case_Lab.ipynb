# Credit Risk Analytics Lab using Synthetic Loan Applicant Data

## Industry Use Case
Banks, NBFCs, lending apps, and FinTech companies use credit risk analytics to decide whether a customer is likely to repay a loan or default.

Credit risk means the possibility that a borrower may fail to repay the loan on time.

## Lab Objective
In this lab, students will learn how to:

1. Create a synthetic loan applicant dataset.
2. Understand real-world credit risk parameters.
3. Calculate Debt-to-Income Ratio (DTI).
4. Classify customers into Low, Medium, and High credit risk groups.
5. Create a loan approval decision based on financial behavior.
6. Identify customers who may default.
7. Visualize credit risk patterns.
8. Understand how banks use AI for credit risk assessment.

## Real-World Relevance
This lab simulates the kind of credit risk workflow used in:

- Loan approval systems
- Credit card approval
- Personal loan screening
- Home loan eligibility checks
- Buy Now Pay Later risk scoring
- FinTech lending apps

## Dataset Note
This dataset is fully synthetic and created only for educational purposes.
It does not contain real customer data.


````
# CELL 1: Import required libraries
# pandas is used for handling tabular data.
# numpy is used for generating synthetic data.
# matplotlib and seaborn are used for visualization.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")
````

````
# CELL 2: Create realistic customer names, cities, professions, and loan types
# These fields help students understand credit risk as a real banking scenario.

np.random.seed(42)

customer_names = [
    "Aarav Sharma", "Priya Mehta", "Rohan Verma", "Neha Gupta", "Karan Malhotra",
    "Ananya Singh", "Vikram Rao", "Sneha Iyer", "Rahul Nair", "Meera Joshi",
    "Arjun Kapoor", "Pooja Sinha", "Nitin Bansal", "Ritika Jain", "Sahil Khan",
    "Divya Menon", "Manish Reddy", "Kavita Das", "Harsh Vardhan", "Isha Agarwal"
]

cities = [
    "Delhi", "Mumbai", "Bengaluru", "Hyderabad", "Pune",
    "Chennai", "Kolkata", "Jaipur", "Lucknow", "Noida"
]

professions = [
    "Salaried Employee", "Business Owner", "Freelancer", "Self-Employed",
    "Government Employee", "Consultant", "Teacher", "Doctor",
    "IT Professional", "Small Trader"
]

loan_types = [
    "Personal Loan", "Home Loan", "Education Loan", "Vehicle Loan", "Business Loan"
]

print("Master lists created successfully.")
````

````
# CELL 3: Generate synthetic credit risk dataset
# We are creating 150 loan applicant records.
# These columns represent real input parameters used in lending decisions.

n = 150

df = pd.DataFrame({
    "Customer_ID": range(2001, 2001 + n),
    "Customer_Name": np.random.choice(customer_names, n),
    "Age": np.random.randint(21, 62, n),
    "City": np.random.choice(cities, n),
    "Profession": np.random.choice(professions, n),
    "Loan_Type": np.random.choice(loan_types, n),
    "Monthly_Income": np.random.randint(25000, 250000, n),
    "Existing_EMI": np.random.randint(0, 90000, n),
    "Loan_Amount_Requested": np.random.randint(100000, 5000000, n),
    "Loan_Tenure_Years": np.random.randint(1, 25, n),
    "Credit_Score": np.random.randint(300, 900, n),
    "Employment_Years": np.random.randint(0, 30, n),
    "Previous_Defaults": np.random.randint(0, 4, n),
    "Missed_EMI_Last_12_Months": np.random.randint(0, 8, n)
})

df.head(10)
````

````
# CELL 4: Calculate important credit risk indicators
# DTI = Existing EMI / Monthly Income
# Loan-to-Income Ratio = Loan Amount Requested / Annual Income
# These ratios help banks understand repayment pressure.

df["DTI"] = (df["Existing_EMI"] / df["Monthly_Income"]).round(2)

df["Annual_Income"] = df["Monthly_Income"] * 12

df["Loan_to_Income_Ratio"] = (df["Loan_Amount_Requested"] / df["Annual_Income"]).round(2)

df[[
    "Customer_Name", "Monthly_Income", "Existing_EMI",
    "Loan_Amount_Requested", "DTI", "Loan_to_Income_Ratio"
]].head(10)
````

````
# CELL 5: Create credit risk classification logic
# This is a simplified rule-based model.
# Real banks use machine learning models, credit bureau data, and internal policies.

credit_risk_level = []
loan_decision = []
interest_rate_category = []
credit_strategy = []
business_outcome = []

for index, row in df.iterrows():
    dti = row["DTI"]
    credit_score = row["Credit_Score"]
    defaults = row["Previous_Defaults"]
    missed_emi = row["Missed_EMI_Last_12_Months"]
    employment_years = row["Employment_Years"]
    lti = row["Loan_to_Income_Ratio"]

    # High Credit Risk:
    # Poor credit score, high debt burden, previous defaults, or high missed EMI count.
    if credit_score < 500 or dti > 0.55 or defaults >= 2 or missed_emi >= 5 or lti > 5:
        credit_risk_level.append("High")
        loan_decision.append("Reject / Manual Review")
        interest_rate_category.append("Very High / Not Eligible")
        credit_strategy.append("Reject loan or ask for guarantor, collateral, or improved credit history.")
        business_outcome.append("High chance of default. Lending may create financial loss.")

    # Medium Credit Risk:
    # Average credit score, moderate debt pressure, or unstable repayment pattern.
    elif credit_score < 680 or dti > 0.35 or defaults == 1 or missed_emi >= 2 or employment_years < 2 or lti > 3:
        credit_risk_level.append("Medium")
        loan_decision.append("Approve with Conditions")
        interest_rate_category.append("Higher Interest Rate")
        credit_strategy.append("Offer smaller loan, higher rate, shorter tenure, or additional verification.")
        business_outcome.append("Customer can be served carefully with controlled lending terms.")

    # Low Credit Risk:
    # Good credit score, stable income, low EMI burden, and no poor repayment history.
    else:
        credit_risk_level.append("Low")
        loan_decision.append("Approve")
        interest_rate_category.append("Standard / Competitive Rate")
        credit_strategy.append("Approve loan with standard terms and competitive interest rate.")
        business_outcome.append("Customer has strong repayment capacity and low default probability.")

df["Credit_Risk_Level"] = credit_risk_level
df["Loan_Decision"] = loan_decision
df["Interest_Rate_Category"] = interest_rate_category
df["Credit_Strategy"] = credit_strategy
df["Business_Outcome"] = business_outcome

df.head(10)
````

````
# CELL 6: View business-friendly credit risk profile
# This table is useful for student interpretation, reports, and presentations.

selected_columns = [
    "Customer_ID", "Customer_Name", "Age", "City", "Profession", "Loan_Type",
    "Monthly_Income", "Existing_EMI", "Loan_Amount_Requested",
    "Credit_Score", "Employment_Years", "Previous_Defaults",
    "Missed_EMI_Last_12_Months", "DTI", "Loan_to_Income_Ratio",
    "Credit_Risk_Level", "Loan_Decision", "Interest_Rate_Category",
    "Credit_Strategy", "Business_Outcome"
]

df[selected_columns].head(10)
````

````
# CELL 7: Count customers by credit risk level
# This gives an overview of the loan applicant portfolio.

risk_distribution = df["Credit_Risk_Level"].value_counts()

print("Credit Risk Distribution:")
print(risk_distribution)
````

````
# CELL 8: Visualize credit risk distribution
# This chart helps management understand the number of risky applicants.

plt.figure(figsize=(7, 5))
sns.countplot(data=df, x="Credit_Risk_Level", order=["Low", "Medium", "High"])
plt.title("Distribution of Credit Risk Levels")
plt.xlabel("Credit Risk Level")
plt.ylabel("Number of Customers")
plt.show()
````

````
# CELL 9: Analyze Credit Score by Credit Risk Level
# Lower credit scores usually indicate higher probability of default.

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Credit_Risk_Level", y="Credit_Score", order=["Low", "Medium", "High"])
plt.title("Credit Score by Credit Risk Level")
plt.xlabel("Credit Risk Level")
plt.ylabel("Credit Score")
plt.show()
````

````
# CELL 10: Analyze DTI by Credit Risk Level
# Higher DTI means the borrower already carries more repayment burden.

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Credit_Risk_Level", y="DTI", order=["Low", "Medium", "High"])
plt.title("Debt-to-Income Ratio by Credit Risk Level")
plt.xlabel("Credit Risk Level")
plt.ylabel("DTI")
plt.show()
````

````
# CELL 11: Analyze missed EMI behavior
# Missed EMI history is a strong signal of repayment risk.

plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df,
    x="Credit_Risk_Level",
    y="Missed_EMI_Last_12_Months",
    order=["Low", "Medium", "High"]
)
plt.title("Missed EMI Count by Credit Risk Level")
plt.xlabel("Credit Risk Level")
plt.ylabel("Missed EMI in Last 12 Months")
plt.show()
````

````
# CELL 12: Loan decision summary
# This shows how many applications are approved, conditionally approved, or rejected.

decision_summary = df["Loan_Decision"].value_counts()

print("Loan Decision Summary:")
print(decision_summary)
````

````
# CELL 13: Visualize loan decision summary
# This is useful for business reporting.

plt.figure(figsize=(8, 5))
sns.countplot(data=df, y="Loan_Decision")
plt.title("Loan Decision Summary")
plt.xlabel("Number of Customers")
plt.ylabel("Loan Decision")
plt.show()
````

````
# CELL 14: Profession-wise credit risk summary
# This helps understand credit risk patterns by customer profession.
# Since this is synthetic data, this should be interpreted only for classroom learning.

profession_summary = df.groupby("Profession").agg(
    Total_Customers=("Customer_ID", "count"),
    Average_Income=("Monthly_Income", "mean"),
    Average_Credit_Score=("Credit_Score", "mean"),
    Average_DTI=("DTI", "mean"),
    Average_Loan_to_Income_Ratio=("Loan_to_Income_Ratio", "mean"),
    Average_Missed_EMI=("Missed_EMI_Last_12_Months", "mean")
).round(2)

profession_summary.sort_values(by="Average_DTI", ascending=False)
````

````
# CELL 15: Loan type-wise credit risk summary
# This helps banks understand which loan types may need more careful screening.

loan_type_summary = df.groupby("Loan_Type").agg(
    Total_Customers=("Customer_ID", "count"),
    Average_Loan_Amount=("Loan_Amount_Requested", "mean"),
    Average_Credit_Score=("Credit_Score", "mean"),
    Average_DTI=("DTI", "mean"),
    Average_Loan_to_Income_Ratio=("Loan_to_Income_Ratio", "mean")
).round(2)

loan_type_summary.sort_values(by="Average_Loan_to_Income_Ratio", ascending=False)
````

````
# CELL 16: Identify high credit risk customers
# These customers need rejection, manual review, collateral, or guarantor.

high_credit_risk = df[df["Credit_Risk_Level"] == "High"]

high_credit_risk[selected_columns].head(10)
````

````
# CELL 17: Identify low credit risk customers
# These customers are more suitable for loan approval.

low_credit_risk = df[df["Credit_Risk_Level"] == "Low"]

low_credit_risk[selected_columns].head(10)
````

````
# CELL 18: Customer-level credit risk explanation function
# Students can change Customer_ID and explain why a customer is risky or safe.

def explain_credit_risk(customer_id):
    customer = df[df["Customer_ID"] == customer_id]

    if customer.empty:
        print("Customer ID not found.")
        return

    row = customer.iloc[0]

    print("Customer Credit Risk Profile")
    print("----------------------------")
    print(f"Customer Name: {row['Customer_Name']}")
    print(f"Age: {row['Age']}")
    print(f"City: {row['City']}")
    print(f"Profession: {row['Profession']}")
    print(f"Loan Type: {row['Loan_Type']}")
    print(f"Monthly Income: ₹{row['Monthly_Income']}")
    print(f"Existing EMI: ₹{row['Existing_EMI']}")
    print(f"Loan Amount Requested: ₹{row['Loan_Amount_Requested']}")
    print(f"Credit Score: {row['Credit_Score']}")
    print(f"Employment Years: {row['Employment_Years']}")
    print(f"Previous Defaults: {row['Previous_Defaults']}")
    print(f"Missed EMI Last 12 Months: {row['Missed_EMI_Last_12_Months']}")
    print(f"DTI: {row['DTI']}")
    print(f"Loan-to-Income Ratio: {row['Loan_to_Income_Ratio']}")
    print(f"Credit Risk Level: {row['Credit_Risk_Level']}")
    print(f"Loan Decision: {row['Loan_Decision']}")
    print(f"Interest Rate Category: {row['Interest_Rate_Category']}")
    print(f"Credit Strategy: {row['Credit_Strategy']}")
    print(f"Business Outcome: {row['Business_Outcome']}")

# Example: Change customer ID to analyze other customers
explain_credit_risk(2001)
````

````
# CELL 19: Export complete credit risk dataset
# Students can use this CSV file in Excel, Power BI, reports, or GitHub submission.

df.to_csv("synthetic_credit_risk_dataset.csv", index=False)

print("Dataset exported successfully as synthetic_credit_risk_dataset.csv")
````
